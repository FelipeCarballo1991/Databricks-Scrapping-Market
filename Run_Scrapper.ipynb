# Run Scrapper

This notebook executes the `scrapper.py` Python file using the `%run` command.

**Why use %run?**

Python files (`.py`) in Databricks cannot be executed directly when they use Spark. They must be run from a notebook using `%run`, which allows the Python file to inherit the notebook's Spark session.

The `scrapper.py` file will:
* Scrape product prices from Coto supermarket
* Store results in `supermercadoetl.bronze.precios_scraping`
* Log execution metrics to audit tables

```python
import sys
import os

# Add Scrapper directory to Python path so imports work
scrapper_dir = os.path.abspath('./Scrapper')
if scrapper_dir not in sys.path:
    sys.path.insert(0, scrapper_dir)

# Execute the scrapper script
exec(open('./Scrapper/scrapper.py').read())
```
